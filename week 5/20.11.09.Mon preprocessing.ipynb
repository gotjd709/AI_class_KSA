```python
import pandas as pd
titanic_df = pd.read_csv("titanic.csv")
titanic_df.head(5)
```

```python
titanic_df.info()
```

```python
titanic_df.drop(columns='Name', inplace=True)
titanic_df.head(5)
```

```python
titanic_df.drop(columns=['Cabin', 'Embarked', 'Ticket'], inplace=True)
# 여러 열을 제거할 때 딕셔너리를 만들어서 제거해야 샤함
titanic_df.head(5)
```

```python
titanic_df.tail(10)
```

```python
titanic_df[titanic_df.duplicated()]
# 중복돈 행의 데이터 자료를 보여줌
```

```python
titanic_df.duplicated()
# 중복된 행을 표시해줌
```

```python
titanic_df.drop_duplicates(inplace = True)
titanic_df[titanic_df.duplicated()]
# 여기까지만 진행하면 컬럼 이름만 나오게 됨
# 중복되 값들을 제거했기 때문
titanic_df.tail(10)
# 제거된 후 데이터를 불러옴
```

```python
titanic_df.isnull().head(10)
```

```python
titanic_df.isnull().sum()
# 누락된 값들이 총 몇 개인지 표시해줌
```

```python
titanic_df.dropna(inplace=True)
# dropna: 누락된 값 제거해주기
titanic_df.isnull().sum()
```

```python
titanic_df.dropna(axis=1)
# axis = 0일 경우 누락된 값의 행을 제거함
# axis = 1일 경우 누락된 값의 열을 제거함
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df.isnull().sum()
```

```python
# columns
# 현재 가지고 있는 데이터의 컬럼 네임을 출력해줌
titanic_df.rename(columns={
    'Survived':'survived', 'PassengerId':'passengerid', 'Pclass':'pclass', 'Gender':'gender', 'Age':'age', 'SibSp':'sibsp', 'Parch':'parch', 'Fare':'fare'
}, inplace = True)
titanic_df.columns
```

```python
titanic_df.replace({'gender':{'male':0}}, inplace=True)
titanic_df.replace({'gender':{'female':1}}, inplace=True)
```

```python
titanic_df.head(10)
```

```python
titanic_df.replace({'survived':{0:'No'}}, inplace=True)
titanic_df.replace({'survived':{1:'Yes'}}, inplace=True)
titanic_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic_df[['fare']] = scaler.fit_transform(titanic_df[['fare']])
titanic_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic_df[['age']] = scaler.fit_transform(titanic_df[['age']])
titanic_df.describe()
# 정규화는 데이터를 스무스하게 해주는 것 '0과 1 사이로'
# 표준화는 데이터릘 말 그대로 표준화 시켜주는 것
```

```python
titanic_df['fare'].plot(kind="box")
# 이상치가 존재하므로 정규화를 했음에도 큰 영향을 줌
```

```python
titanic_df.drop(titanic_df[titanic_df.fare >= 1].index, inplace = True)
titanic_df['fare'].plot(kind="box")
# 제거해 준 후 확인해보기
```

```python
insurance_df = pd.read_csv("insurance.csv")
insurance_df.head(5)
```

```python
insurance_df.drop(columns='Number', inplace=True)
insurance_df.head(5)
```

```python
insurance_df.duplicated()
```

```python
insurance_df.drop_duplicates(inplace = True)
insurance_df.tail(10)
```

```python
insurance_df.dropna(inplace=True)
insurance_df.isnull().sum()
```

```python
insurance_df.rename(columns={
    'Gender':'gender', 'Insurance Fee':'charges'
}, inplace=True)

insurance_df.head(10)
```

```python
insurance_df.replace({'gender':{'female':1}}, inplace=True)
insurance_df.replace({'gender':{'male':0}}, inplace=True)
insurance_df.head(5)
```

```python
insurance_df['age'].plot(kind="box")
```

```python
insurance_df['bmi'].plot(kind="box")
```

```python
insurance_df['charges'].plot(kind="box")
```

```python
insurance_df.drop(insurance_df[insurance_df.charges >= 50000].index, inplace=True)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
insurance_df[['charges']]=scaler.fit_transform(insurance_df[['charges']])
insurance_df.describe()
```

```python
insurance_df.drop(insurance_df[insurance_df.bmi >= 45].index, inplace=True)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
insurance_df[['bmi']]=scaler.fit_transform(insurance_df[['bmi']])
insurance_df.describe()
```

```python
corr=insurance_df.corr(method = 'pearson')
print(corr)
```
