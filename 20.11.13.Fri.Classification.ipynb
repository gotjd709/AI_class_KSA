```python
# kNN model
# k개의 이웃들을 선택하여 많은 수의 이웃이 속해있는 클래스로 분류

import pandas as pd
from sklearn.datasets import load_breast_cancer
breast_cancer_data = load_breast_cancer()
```

```python
breast_cancer_data
```

```python
df_data = pd.DataFrame(breast_cancer_data.data)
df_labels = pd.DataFrame(breast_cancer_data.target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(df_data, df_labels, test_size=0.2, random_state=100) 
```

```python
training_data
```

```python
validation_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
c3 = KNeighborsClassifier(n_neighbors = 3)
c3.fit(training_data, training_labels)
```

```python
print(c3.score(validation_data, validation_labels))
# k=3일때의 결과
```

```python
c1 = KNeighborsClassifier(n_neighbors = 1)
c1.fit(training_data, training_labels)
print(c1.score(validation_data, validation_labels))
# k=1일때의 결과
```

```python
import sklearn
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
export_graphviz(tree, out_file="teee.dot", class_names=["악성", "양성"],
               feature_names=cancer.feature_names, impurity=False, filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(training_data, training_labels)
print(forest.score(training_data, training_labels))
print(forest.score(validation_data, validation_labels))
```

```python
import numpy as np
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1, n_features)
```

```python
plot_feature_importances_cancer(forest)
```
